# RT Testing

### Data Loading

```python
from alphabase.io.hdf import HDF_File
hdf_filepath = "./data_ext/AlphapeptdeepData/hdf/rt/PXD024364/evidence.psm.hdf"
hdf_file = HDF_File(hdf_filepath, read_only=True)
psm_df = hdf_file.psm.psm_df.values
```

```python
from peptdeep.utils import uniform_sampling, evaluate_linear_regression_plot, evaluate_linear_regression
train_df, test_df = uniform_sampling(
    psm_df, target='rt_norm', n_train=10000, return_test_df=True
)
```

### Model_RT_Bert_Local Testing

```python
# Config setting
import warnings
warnings.filterwarnings("ignore")
from peptdeep.model.rt import AlphaRTModel, Model_RT_Bert_Local
```

```python
rtmodel = AlphaRTModel(dropout=0.1, mask_modloss=False, model_class=Model_RT_Bert_Local, output_attentions = True)
```

```python
rtmodel.train(train_df, 
    warmup_epoch=1, epoch=2, batch_size=512, lr=1e-4,
    verbose=True,
    verbose_each_epoch=True
)
```

```python
predict_df = rtmodel.predict(train_df, verbose=True)
```

```python
evaluate_linear_regression_plot(predict_df, x='rt_pred',y='rt_norm')
```

```python
evaluate_linear_regression(predict_df, x='rt_pred',y='rt_norm')
```
