```python
from peptdeep.model.charge import *
import pandas as pd

model = ChargeModelForAASeq(min_charge=1, max_charge=2)

seq_df = pd.DataFrame({
    'sequence': ['ABCDE','FGHIJK','LMNOPQ','RSTUVWXYZ'],
    'charge_indicators': [[1,0],[0,1],[1,1],[0,0]],
})
model.train(seq_df)
model.predict(seq_df)
model.predict_and_clip_charges(seq_df, charge_prob_cutoff=0.3)
```
